```r
subject <- read.table('favoritesujects.csv', header=T, sep=',')
head(subject)
```

```r
library(psych)
library(GPArotation)
subject.factor = principal(subject, nfactors = 3, rotate="none")
names(subject.factor)
subject.factor$loadings
subject.factor$values
# factanal(subject, factors=)
```

```r
subject.Varimax = principal(subject, nfactors=2, rotate="varimax")
subject.Varimax
```

```r
subject.Promax = principal(subject, nfactors=2, rotate="promax")
subject.Promax
```

```r
Fa<-principal(subject, nfactors=2)
Fa$rotation
```

```r
fa(subject, nfactors=2, rotate="Varimax")
```
