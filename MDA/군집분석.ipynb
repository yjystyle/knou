## 4.4 비계층적 군집분석
- 계층적 군집분석은 관찰지의 수가 적은 경우에 적당
- 관찰지의 수가 많은 경우에는 관찰지들 사이의 유사성/거리 행렬을 구하는 것이 매우 번거롭고 방대함
- 예. 관찰치수가 10개인 경우 유사성행렬은 9\*10/2=45 (혹은 1+2+3+...+9=45) 개의 원소를 가지나, 관찰치수가 1000개인 경우는 999\*1000/2=499,500개의 원소를 가짐

- 대용량의 데이터에 적합한 비계층적 군집분석 방법인 **K-평균 군집분석** 사용



### 4.4.1 K평균 군집분석의 절차
1. 군집의 수 K를 정한다.
    * 비계층 군집분석의 가장 큰 단점. 
    * 군집의 수 *K*를 정하기 어렵다.
    * 계층적 군집분석은  *K*를 사전에 정할 필요가 없다.
2. 임의의 *K*개 관찰치를 *K*개 각 군집에 임의로 지정한다. 이를 *K*개 각 군집의 중심으로 이용한다.
3. 모든 관찰치를 군집중심으로 부터 유클라디안 거리가 최소의 군집에 귀속시킨다.
4. 각 군집에 속한 관찰치들을 이용하여 군집중심을 새로 계산한다.
5. 변화(군집간 관찰치이동)가 없을 때까지 단계3과 단계4를 반복한다. 


### 4.4.2 K-평균 분집분석의 예

| :---: | :---: | :---:|
| 1| -1 | 0|
|2|0| 1|
|3|0|-1|
|**4**|**2**|**0**|
|5|3|1|
|6|3|-1|
|**7**|**3**|**0**|

1. 먼저 군집의 수는 *K*=2로 정하기로 한다.
2. 관찰치 4와 7을 임의의 초기값으로 설정한다. 이를 각 군집의 중심점으로 이용한다.
3. 모든 관찰치에 대해서 각 군집 중심점까지의 유클라디안 거리를 측정한다.



$$D(X,Y) = \sqrt{\sum_{i=1}^p (X_i -  Y_i)^2}$$

관찰치 1의 경우를 예로 들어 설명해보자.
1. 군집 1(관찰치 4번)까지의 거리
$$\sqrt{\{2-(-1) \}^2 + (0-0)^2} = 3$$
2. 군집 2(관찰치 7번)까지의 거리
$$\sqrt{\{3-(-1) \}^2 + (0-0)^2} = 4$$
3. *1번*과 *2번*을 비교했을 때 3<4 이므로 *1번* 군집에 할당

- 모든 관찰치에 대해서 새로운 군집 중심점까지의 유클리디안 거리를 측정하면 아래와 같다.

- 그러나, 처음에 선택한 임의의 관찰치 *4(2,0)*과 *7(3,0)*을 기준으로 만든 위의 군집은 관찰치가 변경됨에 따라 군집도 변할 수 있다.

### 4.4.3 기타 고려사항
#### (1) 초기값의 설정
- 처음에 지정된 군집중심(초기값)에 의하여 최종 군집결과에 차이가 발생할 수 있음
- 데이터의 수가 소량일수록 초기값의 설정은 더욱 중요한 문제
- 임의 초기화 : 초기값을 설정하는 대표적 방법은 데이터 내 임의의 K개 관찰치를 각 군집의 초기값으로 설정하는 것
- **계층적 군집분석을 먼저 수행하여 구하여진 K개 군집의 중심점을 K-평균 군집분석의 초기값으로 사용하는 방식도 가능** (계층적 + 비계층적)
- 데이터가 대용량이기 때문에 계층적 군집분석을 사용하여 초기값을 구하기 어려운 경우에는 원데이터로 부터 적은 표본을 추출하여 계층적 군집분석을 수해한 후 초기값을 구할 수도 있음 (계층적 + 비계층적 + 표본추출)
- SAS 혹은 SPSS 등은 초기값을 자동으로 선택

#### (2) 군집의 수 결정
- K-평균 군집분석에서 군집수 K를 얼마로 하느냐에 의하여도 군집결과가 달라질 수 있음
- 군집의 수를 증가시켜가면서 K-평균 군집분석을 반복하여 수행한 후 이러한 결과중 가장 좋은 결과를 보이는 군집의 수를 결정하는 방법 (시행착오 필요)
- 군집분석의 수행 이전에 주성분 분석(차원 축소)을 먼저 수행하고, 상위 2개의 주성분을 이용하여 군집의 개수를 미리 그림(산점도)으로 확인해보는 방법도 있음
- 계층적 군집분석을 먼저 수행하여 군집의 수를 수형도를 통하여 미리 정한 후 비계층적 군집분석을 수행하는 방법

#### (3) 실용적 고려사항
- 빠른 연산으로 인하여 대규모의 데이터에도 손쉽게 군집분석 결과를 구할 수 있다는 장점. 실무적으로 아주 활용도가 높음 (계산량이 적음)
- 검증 : 분석 데이터를 훈련데이터와 검증데이터로 분리하여 각 데이터에 K-평균 군집분석을 수행하고 생성된 K개 군집의 중심점을 비교해 보는 방법
- 검증 : 군집분석의 군집할당 결과를 이용하여 판별분석<sup>[1]</sup>을 수행해 보는 방법이 있다. 판별분석의 결과가 우수하다면 군집분석의 결과를 신뢰할 수 있음


<br>

- 각 변수의 관측단위가 다르기 때문에 각 변수들을 표준화하여 이용하는 것이 바람직함

```r
beer.data = read.table("mvadata/beerbrand.csv", header=T, sep=",")
head(beer.data)
```

```r
summary(beer.data)
```

```r
# Hierarchical Clustering
# Distance Matrix Computation
# 최소연결법 사용
hc = hclust(dist(beer.data), method="single")
hc
```

```r
plot(hc)
```

```r
#  군집 묶어주는 선이 y=0 부터 시작
plot(hc, hang=-1, )
```

# 그러나 표준화를 하지 않았다.........

```r
# install.packages('pls')
library(pls)
# Standardize
zbeer.data=stdize(as.matrix(beer.data))
# 중심연결법
zhc = hclust(dist(zbeer.data), method="centroid")
plot(zhc, hang=-1)
```

```r
zhc.cent24 = cutree(zhc, 2:4)
zhc.cent24
```

```r
# 비계층적 군집분석 (K-means)
kmc = kmeans(zbeer.data,2)
kmc
```

```r
plot(zbeer.data, col=kmc$cluster, pch=16)
```

```r
pairs(zbeer.data, col=kmc$cluster, pch=16)
```
