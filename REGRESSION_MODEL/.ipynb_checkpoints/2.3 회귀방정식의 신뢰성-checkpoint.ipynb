```r
market2 = read.table("reg/market-2.txt", header = T)
market2
```

```r
market2.lm = lm(Y~X1+X2, data = market2)
```

```r
summary(market2.lm)
```

```r
# SS(X1) = 485.57
# SS(X2|X1) = 22.3O4
anova(market2.lm)
```

```r
#R^2 = 0.9799
names(market2.lm)
yhat = market2.lm$fitted
yhat
```

```r
cor(market2$Y, yhat)
```

```r
# R^2
cor(market2$Y, yhat)^2
```
