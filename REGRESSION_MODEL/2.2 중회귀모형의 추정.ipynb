```r
market2 = read.table("reg/market-2.txt", header=T)
head(market2,3)
```

```r
X=market2[,c(2:3)]
```

```r
#   1   ( )
X=cbind(1,X)
```

```r
Y=market2[,4]
Y
```

```r
# X  .
X = as.matrix(X)
X
```

```r
Y = as.matrix(Y)
```

```r
# X'X
XTX = t(X) %*% X
XTX
```

```r
# inverse
XTXI = solve(XTX)
XTXI
```

```r
XTY = t(X) %*% Y
```

```r
beta = XTXI %*% XTY
beta
```

```r
beta = round(beta,3)
beta
```
